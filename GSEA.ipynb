```python
import gseapy
gseapy.get_library_name(database='Mouse') 
```

```python
dict_go = {}
import os
import pandas as pd
from os import walk
import matplotlib.pyplot as plt
for path, subdirs, files in walk(config['PATHS']["OUTPUT"]):
    for f in files:
        if f.startswith("DGE_PT") and f.endswith(".csv"):
            print(os.path.join(path, f))
            data_name = f.split("/")[-1].split(".")[-2]
            print(data_name)
            df = pd.read_csv(os.path.join(path, f))
            df = df.sort_values('qval')
            df = df[df.qval <= 0.05]
            
            print("Used genes: \n",", ".join(df["gene"]))
            print("\n")
            
            enr_res = gseapy.enrichr(gene_list=df["gene"].tolist(),
                                 organism='Mouse',
                                 gene_sets=['GO_Biological_Process_2021'],
                                 description='pathway',
                                 outdir=config['PATHS']["OUTPUT"]+data_name,
                                 cutoff = 0.5)

            display(enr_res.results.head(20))
            #enr_res.results.to_csv(config['PATHS']["OUTPUT"]+data_name+"/DC_"+data_name+".csv")
            dict_go[data_name] = enr_res.results
          
            #Plot Bar and Dotplots
            gseapy.plot.barplot(enr_res.res2d,title=data_name,
                                ofname=config['PATHS']["OUTPUT"]+data_name+"Barplot.pdf")
            gseapy.plot.dotplot(enr_res.res2d, title=data_name,cmap='viridis_r',
                               ofname=config['PATHS']["OUTPUT"]+data_name+"Dotplot.pdf")
            
            gseapy.plot.barplot(enr_res.res2d,title=data_name)
            gseapy.plot.dotplot(enr_res.res2d, title=data_name,cmap='viridis_r')
            plt.show() 
```

## c vs 17

```python
topgo = 10
choosen_go = []
print("included datasets for plot:")
for key in dict_go:
    if key.endswith("_c_17") :
        print(key)
        choosen_go.extend(list(dict_go[key][:topgo].Term.to_list()))

print("\nnumber of choosen GO-Terms")
import numpy as np
choosen_go = np.unique(choosen_go)
print(len(choosen_go))

print("Go-Terms:")
print(choosen_go)
```

```python
import matplotlib.pyplot as plt
from math import pi
N = len(choosen_go)
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# Initialise the spider plot
fig = plt.figure(figsize=(15,15))
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels
plt.xticks(angles[:-1], range(0, len(choosen_go)))#choosen_go)
for label,i in zip(ax.get_xticklabels(),range(0,len(angles[:-1]))):
    angle_rad=angles[:-1][i]
    if angle_rad <= pi/2:
        ha= 'left'
        va= "bottom"
        angle_text=angle_rad*(-180/pi)+90
    elif pi/2 < angle_rad <= pi:
        ha= 'left'
        va= "top"
        angle_text=angle_rad*(-180/pi)+90
    elif pi < angle_rad <= (3*pi/2):
        ha= 'right'
        va= "top"  
        angle_text=angle_rad*(-180/pi)-90
    else:
        ha= 'right'
        va= "bottom"
        angle_text=angle_rad*(-180/pi)-90
    label.set_rotation(angle_text)
    label.set_verticalalignment(va)
    label.set_horizontalalignment(ha)

# Draw ylabels
ax.set_rlabel_position(0)
  
# Plot each individual = each line of the data  
df = pd.DataFrame()
for key in reversed(list(dict_go.keys())):#dict_go:
    if key.endswith("_c_17") :
        df_tmp = pd.DataFrame(data={"categories":choosen_go})
        df_tmp = df_tmp.set_index("categories", drop=False)
        dict_go[key] = dict_go[key].set_index("Term", drop=False)
        df_tmp = pd.concat([df_tmp, dict_go[key][dict_go[key]["Term"].isin(choosen_go)]], axis=1)
        values=df_tmp["Adjusted P-value"].fillna(0).values.tolist()
        values += values[:1]
        values = -np.log10(values)
        ax.plot(angles, values, linewidth=3, linestyle='solid', label=key)
        df[key] = values
        ax.fill(angles, values, 'r', alpha=0.1)

#Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 1))
plt.tight_layout()
#plt.savefig(config['PATHS']["OUTPUT"]+'/spiderplot_c_17.pdf', dpi= 600)
plt.show()

df = pd.DataFrame(data={"GO":choosen_go})
df[["GO", "GO-ID"]] = df.GO.str.split(" \(", expand=True)
df["GO-ID"] = df["GO-ID"].str.strip("\)")
pd.set_option('display.max_colwidth', None)
df
```

```python
key = "DGE_PTsc_c_17"
df_tmp_sc = dict_go[key][dict_go[key]["Term"].isin(choosen_go)][["Term","Adjusted P-value", "Overlap"]]
df_tmp_sc["-log10 Adjusted P-value"] = -np.log10(df_tmp_sc["Adjusted P-value"])
df_tmp_sc["seq"] = "sc"

key = "DGE_PTsn_c_17"
df_tmp_sn = dict_go[key][dict_go[key]["Term"].isin(choosen_go)][["Term","Adjusted P-value", "Overlap"]]
df_tmp_sn["-log10 Adjusted P-value"] = -np.log10(df_tmp_sn["Adjusted P-value"])
df_tmp_sn["seq"] = "sn"

import seaborn as sb
palette = ['orange', "blue"]
fig = sb.scatterplot(data=df_tmp_sn.append(df_tmp_sc),x="-log10 Adjusted P-value", y="Term",
                     hue="seq", palette=palette).set_title("c vs. 17")
plt.legend(loc = 2, bbox_to_anchor = (1,1))
plt.tight_layout()
#plt.savefig(config['PATHS']["OUTPUT"]+'/figs/dotplot_go_c_17.pdf',dpi= 600,  bbox_inches = "tight")
plt.show()
```
